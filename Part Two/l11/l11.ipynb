```python
#2
from matplotlib import image, pyplot
import os

path = os.path.dirname(os.path.abspath(__file__))
filename = path + "\\..\\..\\samples\\lenna.bmp"
filename2 = path + "\\usa.png"
lenna_data = image.imread(filename)
flag_data = image.imread(filename2, "png")
flag_dimensions = flag_data.shape

plot_data = lenna_data.copy()
for width in range(flag_dimensions[1]):
    for height in range(flag_dimensions[0]):
        plot_data[height][width + 256][0] = flag_data[height][width][0] * 255
        plot_data[height][width + 256][1] = flag_data[height][width][1] * 255
        plot_data[height][width + 256][2] = flag_data[height][width][2] * 255

# Write the modified images
image.imsave(path+'/'+'lenna-mod.jpg', plot_data)

pyplot.imshow(plot_data)
pyplot.show()
```

```python
#3
import PyPDF2

file_handle = open('Sense-and-Sensibility-by-Jane-Austen.pdf', 'rb') 
pdfReader = PyPDF2.PdfReader(file_handle) 
page_number = len(pdfReader.pages)   # this tells you total pages
frequency_table = {}

for page_object in pdfReader.pages:
    page_text = page_object.extract_text()   # this is the str type of full page
    lines = page_text.split("\n")
    for line in lines:
        line = line.lower()
        if line.startswith("chapter") or line.startswith("page"):
            continue
        words = line.split()
        for word in words:
            word = word.strip(",.!@#$%^&*();")
            if word.isalpha():
                if word in frequency_table:
                    frequency_table[word] += 1
                else:
                    frequency_table[word] = 1

print(frequency_table)
```
